```python
def student_creation():
    stopper = ''
    while stopper != 'a':
        stopper = input(' To create a new character data press 1 \n To exit press a \n')
        if stopper == '1':
            id = input('Id of student : ')
            name = input('Name of student : ')
            sirname = input('Second name of student : ')
            fathername = input('Father` name of student : ')
            mothername = input('Mother`s name of student : ')
            age = input('Age of student : ')
            siblings = input('Number of siblings : ')
            s_class = input('Class of student : ')
            with open('student data.csv', 'a', newline = "\n") as f:
                data_writer = csv.writer(f, delimiter = ',')
                data_writer.writerow([id, name, sirname, fathername, mothername, age, siblings, s_class])
```

```python
def student_read():
    stopper = ''
    while stopper != 'a':
        stopper = input('To search via id press 1\nTo search via name press 2\nTo search via age press 3\nTo exit press a\n')
        if stopper == '1':
            id = input("What id do you want to print ")
            data = []
            with open('student data.csv', 'r') as f:
                data_read = csv.reader(f)
                for i in data_read:
                    data.append(i)
            for i in data:
                    for j in i:
                        if (j == id):
                            print(i)
        elif stopper == '2':
            name = input("What name do you want to print ")
            data = []
            with open('student data.csv', 'r') as f:
                data_read = csv.reader(f)
                for i in data_read:
                    data.append(i)
            for i in data:
                    for j in i:
                        if (j == name):
                            print(i)
        elif stopper == '3':
            age = input("What age do you want to print ")
            data = []
            with open('student data.csv', 'r') as f:
                data_read = csv.reader(f)
                for i in data_read:
                    data.append(i)
            for i in data:
                    for j in i:
                        if (j == age):
                            print(i)
```

```python
def student_update():
    stopper = ''
    while stopper != 'a':
        stopper = input('Press 1 to change student data\nPress a to exit')
        if stopper == '1':
            change = input('What id data do you want to change')
            data1 = []
            with open('student data.csv', 'r') as f:
                data_read = csv.reader(f)
                for i in data_read:
                    data1.append(i)
            for i in data1:
                    for j in i:
                        if (j == change):
                            print(i)

            index = int(input('Press 0 to change id\nPress 1 to change name\nPress 2 to change sir name\nPress 3 to change father name\nPress 4 to change Mother name'))
            changer = input('What is the new value ')
            data2 = []
            with open('student data.csv', 'r') as f:
                data_reader = csv.reader(f)
                for i in data_reader:
                    data2.append(i)
            for i in data2:
                for j in i:
                    if (j == change):
                        i[index] = changer
                        print(i)
            with open('student data.csv','w') as f:
                data_writer = csv.writer(f, delimiter = ",")
                data_writer.writerows(data2)
```

```python
def student_delete():
    stopper = ''
    while stopper != 'a':
        stopper = input('To delete a character data press 1 \n To exit press a \n')
        if stopper == '1':
            delete = input('Enter the character id you want to delete ')
            data3 = []
            with open('student data.csv', 'r') as f:
                data_read = csv.reader(f)
                for i in data_read:
                        data3.append(i)
                for i in data3:
                        for j in i:
                            if (j == delete):
                                print(i)
            warning = input('Are you sure you want to delete it ')
            if warning == 'Yes' or 'yes':
                data4 = []
                with open('student data.csv', 'r') as f:
                    data_reader = csv.reader(f)
                    for i in data_reader:
                        data4.append(i)
                for i in data4:
                    for j in i:
                        if (j == delete):
                            del data4[i]
                with open('student data.csv','w') as f:
                    data_writer = csv.writer(f, delimiter = ",")
                    data_writer.writerows(data4)
            elif warning != 'Yes' or 'yes':
                print('Noting is deleated')
                
```

```python
def school_management():
    stopper = ''
    while stopper != 'q':
        stopper = input(' To create a new student data press 1 \n To update student data press 2 \n To read student data press 3 \n To delete a student data press 4 \n To exit the program press q \n')
        if stopper == '1':
            student_creation()
        elif stopper == '2':
            student_update()
        elif stopper == '3':
            student_read()
        elif stopper == '4':
            student_delete()
```

```python
school_management()
```
