```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('test.xlsx')
```

```python
data_f = data.reindex(['Emp N0','Name', 'History', 'Date'], axis = 1)
```

```python
newdata = pd.DataFrame()
newdata['Per'] = ''
newdata['Per'] = data_f['Emp N0'].unique()
newdata['Name'] = ''
newdata['Name'] = data_f['Name'].unique()
```

```python
newdata.drop('Date of Assinment', axis = 1, inplace=True)
```

```python
new_c0 = []
for i in newdata['Per']:
    new_c0.append(data_f[(data_f['Emp N0'] == i) & (data_f['History'] == 'Date of Join')]['Date'].iloc[0])
newdata['Data of Join'] = new_c0
```

```python
new_c1 = []
for i in newdata['Per']:
    new_c1.append(data_f[(data_f['Emp N0'] == i) & (data_f['History'] == 'Date of Assinment')]['Date'].iloc[0])
newdata['Data of Assinment'] = new_c1
```

```python
new_c2 = []
for i in newdata['Per']:
    new_c2.append(data_f[(data_f['Emp N0'] == i) & (data_f['History'] == 'Date of last Appraisal')]['Date'].iloc[0])
newdata['Data of last Appraisal'] = new_c2
```

```python
new_c3 = []
for i in newdata['Per']:
    new_c3.append(data_f[(data_f['Emp N0'] == i) & (data_f['History'] == 'Date of Termination')]['Date'].iloc[0])
newdata['Data of Termination'] = new_c3
```

```python
newdata.to_excel('Final Data_Sorted.xlsx')
```
